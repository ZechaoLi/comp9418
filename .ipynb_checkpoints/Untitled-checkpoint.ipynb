```python

import zipfile
import numpy as np
from sklearn.decomposition import SparsePCA
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from scipy.sparse import lil_matrix

def read_zip(filename):

    zip_f = zipfile.ZipFile(filename)
    #print(zip_f.namelist())
    file_name_list = zip_f.namelist()[1:]
    x_file = []
    sub_x = []
    y_i = 0
    data_x = None
    data_y = None
    all_x = []
    
    for file in file_name_list:
        if file[-1] == 'x':
            x_file.append(file)

    for xfile in x_file:
        temp_x_dict = dict()
        file_x_c = zip_f.read(xfile).decode('utf-8').split("\n")[:-1]
        file_y_c = [int(e) for e in zip_f.read(xfile[:-2] + ".y").decode('utf-8').split("\n")[:-1]]

        
        for e in file_x_c:
            l = e.split(' ')
            if int(l[0]) not in temp_x_dict:
                temp_x_dict[int(l[0])] = [int(l[1])]
            else:
                temp_x_dict[int(l[0])].append(int(l[1]))
        all_x.append(temp_x_dict)
        #print(temp_x_dict)
        #sub_x = np.zeros(shape= [len(temp_x_dict), 2035523])
        for e in temp_x_dict:
            sub_x.append(e)
        #    for i in temp_x_dict[e]:
        #        sub_x[e -1][i - 1] = 1

        #if x_i == 0:
        #    data_x = sub_x
        #    x_i = 1
        #else:
        #    data_x = np.vstack((data_x, sub_x))


        sub_y_lable = np.zeros(shape = [len(file_y_c), 1])
        for i in range(len(file_y_c)):
            sub_y_lable[i][0] = file_y_c[i]
        if y_i == 0:
            data_y = sub_y_lable
            y_i = 1
        else:
            data_y = np.vstack((data_y, sub_y_lable))


    #data_x = np.zeros(shape=[len(sub_x), 88441])
    data_x = lil_matrix((len(sub_x), 88441))
    print(len(all_x))
    k = 0
    for x in all_x:
        for e in x:
            for i in x[e]:
                data_x[k,i - 1] = 1
            k += 1
    #print(data_x.shape, data_y.shape)
    '''
    data_x = np.zeros(shape=[len(sub_x), 19])
    for e in temp_x_dict:
        for i in range(len(temp_x_dict[e])):
            data_x[e - 1][i] = temp_x_dict[e][i]
    '''
    return data_x, data_y
```
